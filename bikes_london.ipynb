```python
new_cols_dict ={
    'timestamp':'tiempo',
    'cnt':'enumerar', 
    't1':'temp_C',
    't2':'sensacion_termica_C',
    'hum':'porcentaje_humedad',
    'wind_speed':'velocidad_viento_kmph',
    'weather_code':'clima',
    'is_holiday':'feriado',
    'is_weekend':'fin_de_semana',
    'season':'estacion'
}

df.rename(new_cols_dict, axis=1, inplace=True)
```

```python
df.porcentaje_humedad = df.porcentaje_humedad / 100
```

```python
season_dict = {
    '0.0':'primavera',
    '1.0':'verano',
    '2.0':'otoño',
    '3.0':'invierno'
}

weather_dict = {
    '1.0':'Despejado',
    '2.0':'Nubes Dispersas',
    '3.0':'Mayormente nublado',
    '4.0':'Nublado',
    '7.0':'Lluvia',
    '10.0':'Tormenta',
    '26.0':'Nieve'
}

df.estacion = df.estacion.astype('str')

df.estacion = df.estacion.map(season_dict)


df.clima = df.clima.astype('str')

df.clima = df.clima.map(weather_dict)
```
